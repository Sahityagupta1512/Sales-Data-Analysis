```python
#2. Cleaning the dataset.

df.isnull().sum()
```

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

```python
#Inputing missing values for Postal Code

df['Postal Code'].fillna(-1, inplace=True)
```

```python
df.isnull().sum()
```

```python
# Convert 'Order Date' and 'Ship Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()

duplicates
```

```python
# Sales distribution by Category
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

# Plotting sales by category using matplotlib
plt.figure(figsize=(8, 5))
plt.bar(category_sales.index, category_sales.values)
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

<h3> 1. The Highest Sales according to the category is Technology. </h3>
<h3> 2. The Second Highest Sales is Furniture. </h3>
<h3> 3. The Third Highest Sales is Office Suppliers. </h3>

```python
subcategory_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

#Plotting sales by Sub-Category Products
plt.figure(figsize=(5,8)) 
plt.bar(subcategory_sales.index, subcategory_sales.values)
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

<h3> According to my analysis the highest sales according to sub category was Phones </h3>

```python
State_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False)

# Plotting sales by State in USA
plt.figure(figsize=(8, 5))
plt.bar(State_sales.index, State_sales.values)
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

<h3> According my analysis Califronia is the highest sales state. </h3>

```python
Ship_mode = df.groupby('Ship Mode')['Sales'].sum().sort_values(ascending=False)

# Pie chart, where the slices will be ordered and plotted counter-clockwise
labels = Ship_mode.index
sizes = Ship_mode.values
colors = ['skyblue', 'orange', 'green','pink']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution by Ship Mode')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

<h3> The maximum ship mode used by consumer was standard class according to my analysis.

```python
# Total Sales Over Time
df['month'] = df['Ship Date'].dt.to_period('M')
monthly_sales = df.groupby('month')['Sales'].sum()

monthly_sales.plot(kind='line')
plt.title('Monthly Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

<h3> The highest sales was in Year 2018 according to my analysis

```python
region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

# Plotting sales by State in USA
plt.figure(figsize=(8, 5))
plt.bar(region.index, region.values)
plt.title('Total of Ship Modes')
plt.xlabel('Ship Mode')
plt.ylabel('No. of Consumers')
plt.xticks(rotation=90)
plt.show()
```
